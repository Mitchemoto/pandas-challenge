## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
#unique count (len)
unique_df=len(pymoli_df['Item Name'].unique())
unique_df
#average purchase price
avg_price_df=round(pymoli_df['Price'].mean(),2)
avg_price_df
#total purchase value
rev_df=pymoli_df['Price'].sum()
rev_df
#throw into a data frame to return all values
cleaned_df=pd.DataFrame({
        "Number of Unique Items":[unique_df],
        "Average Price":["$"+str(avg_price_df)],
        "Number of Purchases":[count_df],
        "Total Revenue":["$"+str(rev_df)]
        })
cleaned_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Use value counts to get count of total users of specific gender
gender_counts=pymoli_df["Gender"].value_counts()

#formula for percentage of players
gender_percent=gender_counts/count_df*100

#create the data frame 
gender_val_df=pd.DataFrame({
            "Player Percent":gender_percent,
            "Total Count":gender_counts
            })

#format to percentage
pd.options.display.float_format = '{:.2f}%'.format

#run the data frame outside to capture formatting
gender_val_df.head()
```

```python
#groupby gender total count of purchases
gender_count=pymoli_df.groupby(['Gender']).count()['Price']
#Average purchage price
gender_avg_purchase=pymoli_df.groupby(['Gender']).mean()['Price']
#total purchase value
gender_total=pymoli_df.groupby(['Gender']).sum()['Price']
#average total purchase per person 
#issues getting to by person, review back
gender_avg_person=gender_total/gender_val_df['Total Count']

#create the data frame
gender_purchase_df=pd.DataFrame({
            'Purchase Count':gender_count,
            'Average Purchase Price':gender_avg_purchase,
            'Total Purchase Value':gender_total,
            'Avg Total Purchase per Person':gender_avg_person
            })

#update format to show $ value over % value thats being carried over. 
pd.options.display.float_format = '${:.2f}'.format

#run the data frame outside to capture formatting
gender_purchase_df.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create the bins for ages
age_bin=[0,9.90,14.90,19.90,24.90,29.90,34.90,39.90,999]

#title index
age_index=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#organize players age bins by using pd.cut
players_age_df=pymoli_df
players_age_df['Age Range']=pd.cut(players_age_df['Age'],age_bin,labels=age_index)

#create a df based off new groupings by age
players_summary_df=players_age_df.groupby('Age Range')
players_summary_df.count()

#calclulate the averages
summary_df=pd.DataFrame(players_summary_df.count())
summary_df["Purchase ID"]=summary_df["Purchase ID"]/count_df*100

#reduce the columns to two, then create the correct headers to produce the dataframe
reduced_df=summary_df[['SN','Purchase ID']]
final_df=reduced_df.rename(columns={'SN':'Total Count','Purchase ID':"Percentage of Players"})


#holding format from previous kernel. leave be for now
#format to percentage
pd.options.display.float_format = '{:.2f}%'.format

#print the data frame
final_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#keep working with bin/title dataframe - format created in previous kernel and carry over
players_age_df['Age Range']=pd.cut(players_age_df['Age'],age_bin,labels=age_index)

#purchase count by age range
age_count=players_age_df.groupby(['Age Range']).count()['Price']

#Average purchage price
age_avg_purchase=players_age_df.groupby(['Age Range']).mean()['Price']

#total purchase value
age_total=players_age_df.groupby(['Age Range']).sum()['Price']

#average total purchase per person 
#issues getting to by person, review back the 
age_avg_person=age_total/age_count

#create the data frame
age_purchase_df=pd.DataFrame({
            'Purchase Count':age_count,
            'Average Purchase Price': age_avg_purchase,
            'Total Purchase Value':age_total,
            'Avg Total Purchase per Person':age_avg_person
            })

#update format to show $ value over % value thats being carried over. 
#update formatting to each variable, not recognizing count as integer
pd.options.display.float_format = '${:.2f}'.format

#print the dataframe
age_purchase_df
```

## Top Spenders

```python
#Get count of purchases by 'SN'
sn_counts=pymoli_df.groupby(['SN']).count()['Price']
#Average purchage price
sn_avg_purchase=pymoli_df.groupby(['SN']).mean()['Price']
# #total purchase value
sn_total=pymoli_df.groupby(['SN']).sum()['Price']

#create the data frame
sn_purchase_df=pd.DataFrame({
            'Purchase Count':sn_counts,
            'Average Purchase Price':sn_avg_purchase,
            'Total Purchase Value':sn_total,
            })
#sort to show most purchases first
sn_sort=sn_purchase_df.sort_values("Total Purchase Value",ascending=False)

#update format to show $ value over % value thats being carried over. 
pd.options.display.float_format = '${:.2f}'.format

#run the data frame outside to capture formatting
sn_sort.head()
```
